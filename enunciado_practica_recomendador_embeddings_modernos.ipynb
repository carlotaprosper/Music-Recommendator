Once the dataset doesn´t have any missing values or duplicates, we are going to apply two functions to ensure the data is really clean. For that beautifulsoup cleans text removing html tags, the regex to clean any breaklines, tabs... and language detection with langdetect to ensure all summaries and reviews are on the same language. This is very important because most of the models work with english.

Note: the cleaning and language detection is done before the chunking because when we did it the other way the cleaning took way longer and the langdetect didn´t work with small chunks of words, it said spanish when it was english...

```python
# pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
import re

def clean_html_and_text(text):
    if not isinstance(text, str):
        return ""
    
    # replace literal break lines (as characters) with spaces
    clean_text = text.replace('\n', ' ').replace('\t', ' ').replace('\r', ' ').replace('\\', ' ')
    
    # change the break lines with a space
    clean_text = re.sub(r'[\n\r\t]+', ' ', text)
    
    # remove all the HTML tags (<br>, <i>...) and replaces for example &amp; in &
    clean_text = BeautifulSoup(clean_text, "html.parser").get_text(separator=" ")

    # so there´s spaces between words and characters such as , or ;
    clean_text = re.sub(r'([,;:!?])([a-zA-Z])', r'\1 \2', clean_text)
    
    # remove break lines and multiple blank spaces with Regex
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    
    return clean_text

music['summary'] = music['summary'].apply(clean_html_and_text)
music['review'] = music['review'].apply(clean_html_and_text)
music.sort_index(ascending=True)
```

Looking at the df display we thougth the \n were errors, turns out it just pandas way of showing yoy there´s a break line there, so now the df is defenitly clean. We´re good to continue with the language detection.

```python
from langdetect import detect
from langdetect.lang_detect_exception import LangDetectException

def detect_language(text):
    try:
        # detects language
        return detect(text)
    except LangDetectException:
        return "unknown"

music['language'] = music['review'].apply(detect_language)

print(music['language'].value_counts())
```

After trying our first chunking we realized that some data are fill with Not Available, so we want to drop those album or artist names that are not available (there´s no recommendation w/o the artist/album name) before doing the final chunking. Also for the rest of the columns we are going to replace the not avilabe with Unknown, so the embedding undersantds better.

```python
# we save the ones in english for the embedding models
music = music[music['language'] == 'en'].copy()

# drop Not available artists or albums
music = music[music["artist"].str.contains("Not Available") == False]
music = music[music["album"].str.contains("Not Available") == False]
# replace the left not availables with Unkowns
music = music.replace('Not Available', 'Unknown')
display(music.sort_index(ascending=False))
```

Now that´s done, we are going to deep into the mean length of the reviews to see if it reaches the token limit, and if so think about chunking, truncating or just saving some parts of the texts.

```python
meanwords_sum = 0
for row in music['summary']:
    words = row.split()
    meanwords_sum += len(words)

meanwords_rev = 0
for row in music['review']:
    words = row.split()
    meanwords_rev += len(words)

print(f"Mean text length for summary column: {meanwords_sum / 25708}")
print(f"Mean text length for review column: {meanwords_rev / 25708}")
```

In advanced of the embedding, as the normal amount of tokens used in standard multilingual embeddings or the majority models are of 512, we are going to chunk the reviews so they don´t pass the token limit. There was the option of grabbing just the first paragraph but we realized that frequently this part of the review was about stories and not about the album itself, so it wouldn´t be really informative about the albumns. Also, we don´t exactlty know in each review were is the interesting part, so we don´t want to choose a part of the text that can be the one with no relevant information.

```python
# function fot the chunking, chunks of 300 to be more efficient as the token is of 512
def chunk_text(text, chunk_size=300):
    # in case there´s a null
    if not isinstance(text, str):
        return []
    
    words = text.split()
    
    # group by sets of words with a 'chunk_size'
    chunks = [' '.join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size)]
    return chunks

music['summary_chunks'] = music['summary'].apply(chunk_text)
music['review_chunks'] = music['review'].apply(chunk_text)

# chunks per album
music['num_chunks_sum'] = music['summary_chunks'].apply(len)
music['num_chunks_review'] = music['review_chunks'].apply(len)
display(music[['artist', 'album', 'num_chunks_sum', 'num_chunks_review']].sort_values(by='num_chunks_review'))
```

As it is shown, the summary columns have really short texts so the chunks are the whole text, while in the reviews the texts are so long some of them need more than 10 chunks of 300 words.

After dealing with the all the info cleaning, we´re going to analyse the more interesting metadata, like the genre and score distribution.

```python
music = music.fillna('Unknown')

genre_counts = music['genre'].value_counts(dropna=False)
frequency = genre_counts[genre_counts > 100]

print(f"Genre frequency: {frequency}")
```

As it is clear, there´s mixed genres and the df puts them as 1, so we´re going to separate them in lists, so it is clear how many different genres are there.

```python
import re

def genre_to_list(raw_genre):
    text = str(raw_genre).strip()
    
    if text in ['Not Available', 'nan', 'None', '']:
        return ['Unknown']
        
    # remove any kind of unwanted character
    text = re.sub(r'[\[\]\'\"\\+]', '', text)
    
    text = text.replace('/', ',')
    
    # split by , and save the genres
    clean_list = [g.strip() for g in text.split(',') if g.strip()]

    if not clean_list:
        return ['Unknown']
        
    return clean_list

music['genre'] = music['genre'].apply(genre_to_list)

display(music.sort_index(ascending=True))
```

### Attempt A: Structural way, with labeling

We´re gonna add variables like the artist, album and genre to the start of the review chunks using separators like |, like Artist: Radiohead | Album: Kid A | Genre: Electronic, Rock | Score: 10.0 | Review: The synths in the opening track...

This is a really forward method that doesn´t add a lot of length to the chunks which is perfect for the token limit. But, sometimes, models can be confused by so many separators like | or :, so we will see.

### Attempt B: Natural language

Here we´re going to redact a short human presentation/introduction on the start of its review. Embedding models usually understand this attempt better. Example: The album 'Kid A' by Radiohead, is an Electronic and Rock album rated with a 10.0 out of 10.0 score, reviewer said: The synths in the opening track... Model would get the semantic relation in a better way, but there´s a probability of the token limit being reached (maybe not).

### Attempt C: Reviews with just the artist and album

Sometimes the score can introduce noise to the model, because it doesn´t understand that 9 is better score than 8, it just reads the texts. Example, Radiohead - Kid A. The synths in the opening track... This attempt avoids noise.

```python
music_final = music_final.sort_index(ascending=True)

def representations(row):
    # differenet variants
    structured = []
    nl = []
    wo_noise = []
    
    # metadata to be used
    artist = str(row['artist']).title()
    album = str(row['album']).title()
    score = str(row['score'])

    # join the different genres
    genres = ", ".join(row['genre'])

    chunks = row['review_chunks']

    # structures with the chunks in different representations
    for chunk in chunks:
        phrase_A = 'Artist: ' + artist + ' | Album: ' + album + ' | Genre: ' + genres + ' | Score: ' + score + ' | Review: ' + chunk
        structured.append(phrase_A)

        phrase_B = 'The album ' + album + ' by ' + artist + ' is a ' + genres + ' record rated with a ' + score + ' out of 10 score. Reviewer said: ' + chunk
        nl.append(phrase_B)

        phrase_C = artist + ' - ' + album + '. ' + chunk
        wo_noise.append(phrase_C)

    return pd.Series([structured, nl, wo_noise])

# new data frame with final chunks
music_repres = music_final[['artist', 'album']].copy()
music_repres[['chunks_var_A', 'chunks_var_B', 'chunks_var_C']] = music_final.apply(representations, axis=1)

display(music_repres)
```

Now we´ve done the different variants let´s see which one has a better performace, based on 3 things:
1. Visual analysis: Attempt B (Natural language) is the most interpretable onem it is read as a normal/real sentence, so every person/model undersantds perfectly the context/meaning. On the other hand, attempt A (Structured) is easier to read by humnas because the information is structured as a summary, but for the models the symbol characters could break the normal phrase structured.

2. The one that introduces most noise: attempts A y B are risky because they add the score, so there´s a probability that the model starts to group the records by score instead of grouping them by similar reviews. And, attempt A, constantly repeats words like artist: or album:... in each chunk, this introduces unuserful tokens without introducing significant info. So, noise wise, the best option is attempt C. It only gives minimun context of who the artist is and the musical description.

3. The one that seems to work better: this would be solved in a quickly experiment on posterior parts.

## Posibles estrategias

### Estrategia 1 — Ítem a ítem
- obtenéis embeddings de todos los ítems,
